```python
# FILTERING :
# Q1, Get me the employees details who are having salary more than 50K and they belongs to
# deptno = 10
# SQL : select * from emp where dept_no=10 and salary>50000
# way 1:
import pandas as pd
df = pd.read_csv("venv/emp_data.csv")
dept_no_10 =df['deptno']==10
sal_50000=df['salary']>50000
final_filter = dept_no_10 & sal_50000
# way 2
df[(df['deptno'] == 10) & (df['salary'] > 50000)]
```

```python
# Q2. Get me all the employees who belongs to either deptno 10 or deptno 20
# way 1:
df = pd.read_csv("venv/emp_data.csv")
dept_no_10 = df['deptno']==10
dept_no_20 = df['deptno']==20
final_filter = dept_no_10 | dept_no_20
```

```python
df[final_filter]
```

```python
# Q3. Get me all the employees who do not belongs to deptno 10
# SQL : select * from emp where dept_no !=10
# way 1:
df = pd.read_csv("venv/emp_data.csv")
dept_no_not_10 = df['deptno'] == 10
dept_no_not_10 = ~dept_no_10
df[dept_no_not_10]
```

```python
# Sorting ( similar to order by clause in SQL )
# sort by value ( sort_value()), by default in ascending order
# Q1. Sort the data in ascending order of depnto
df.sort_values(by='deptno',ascending=True)
```

```python
# Q1. Sort the data in descending order of depnto
df.sort_values(by='deptno',ascending=False)
```

```python
# Sort the data based on multiple columns
# Q1. Sort the data in ascending order of depnto and descending order of salary
df.sort_values(by=['deptno','salary'],ascending=[True,False])
```

```python
# Sorting by index
df.sort_index(ascending=False)
```

```python
# Sort the data based on multiple columns and persist in the dataframe ( original dataframe)
# Q1. Sort the data in ascending order of depnto and descending order of salary and save in the dataframe
df.sort_values(by=['deptno','salary'],ascending=[True,False],inplace=True)
```
